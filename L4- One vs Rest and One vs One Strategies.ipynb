PRACTICAL-4 : Train and evaluate a multiclass classification model, implement One-vs-Rest and One-vs-One strategies, compare models using F1 score and accuracy.

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.metrics import accuracy_score, f1_score
```

```python
digits = load_digits()
X = digits.data
y = digits.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
base_model = LogisticRegression(max_iter=1000)
```

```python
ovr_model = OneVsRestClassifier(base_model)
ovr_model.fit(X_train, y_train)

y_pred_ovr = ovr_model.predict(X_test)

ovr_acc = accuracy_score(y_test, y_pred_ovr)
ovr_f1 = f1_score(y_test, y_pred_ovr, average='weighted')

print("OvR Performance:")
print(f"Accuracy: {ovr_acc:.4f}")
print(f"F1 Score: {ovr_f1:.4f}")
```

```python
ovo_model = OneVsOneClassifier(base_model)
ovo_model.fit(X_train, y_train)

y_pred_ovo = ovo_model.predict(X_test)

ovo_acc = accuracy_score(y_test, y_pred_ovo)
ovo_f1 = f1_score(y_test, y_pred_ovo, average='weighted')

print("\nOvO Performance:")
print(f"Accuracy: {ovo_acc:.4f}")
print(f"F1 Score: {ovo_f1:.4f}")
```

```python
# Evaluation
print("OvR:")
print("Accuracy:", accuracy_score(y_test, y_pred_ovr))
print("F1 Score:", f1_score(y_test, y_pred_ovr, average='weighted'))

print("\nOvO:")
print("Accuracy:", accuracy_score(y_test, y_pred_ovo))
print("F1 Score:", f1_score(y_test, y_pred_ovo, average='weighted'))
```
